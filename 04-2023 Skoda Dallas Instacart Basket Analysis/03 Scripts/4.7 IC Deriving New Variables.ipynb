## 4.7 Exercise Instructions

```python
#Creating subset of dataframe to avoid processing issues
df = df_ords_prods_merged[:1000000]
```

```python
df.shape
```

```python
#If statement to define different price ranges
def price_label(row):

  if row['prices'] <= 5:
    return 'Low-range product'
  elif (row['prices'] > 5) and (row['prices'] <= 15):
    return 'Mid-range product'
  elif row['prices'] > 15:
    return 'High range'
  else: return 'Not enough data'
```

```python
#Applying the function
df['price_range'] = df.apply(price_label, axis=1)
```

```python
#Values in the new function
df['price_range'].value_counts(dropna = False)
```

```python
#Checking the maxmimum value of the subset dataframe
df['prices'].max()
```

```python
#Using the loc() function to refine original funtion; If prices are greater than 15
df.loc[df['prices'] > 15, 'price_range_loc'] = 'High-range product'
```

```python
#If prices are greater than 5 and less than or equal to 15
df.loc[(df['prices'] <= 15) & (df['prices'] > 5), 'price_range_loc'] = 'Mid-range product' 
```

```python
#If prices are less than or equal to 5
df.loc[df['prices'] <= 5, 'price_range_loc'] = 'Low-range product'
```

```python
#Values in the function
df['price_range_loc'].value_counts(dropna = False)
```

### Observations: Anything above 2,556,034.5 is top, which includes hours 10, 11, 14, 15, 13 and 12.  Anything below 272,562 is bottom, whih includes hours 3, 4, 2, 5, 1 and 0.  Anything else is in the middle

```python
#Create a for loop to separate values based on above quartiles
busy_hours = []
```

```python
for hour in df_ords_prods_merged['order_hour_of_day']:
  if hour in [10, 11, 14, 15, 13, 12]:
    busy_hours.append('Most orders')
  elif hour in [0, 1, 5, 2, 4, 3]:
    busy_hours.append('Fewest orders')
  else:
    busy_hours.append('Average orders')
```

```python
#Create the busiest_period_of_day column
df_ords_prods_merged['busiest_period_of_day'] = busy_hours
```
