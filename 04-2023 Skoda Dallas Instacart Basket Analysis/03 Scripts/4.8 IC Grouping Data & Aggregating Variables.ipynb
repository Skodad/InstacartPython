### Question 7. Determine Frequent vs Non-Frequent customers.  Create an order frequency flag that marks the regularity of a user's ordering behavior according to the median in the days_since_prior_order column.  Criteria as follows:

### If the median is higher than 20, customer should be labeled "Non-Frequent customer"

### If the median is higher than 10 and lower than of equal to 20, then the customer should be labeled "Regular customer

### If the median is lower than 10, then the customer should be labeled a "Frequent customer"

```python
#Split the data into groups based on user_id column and apply the transform function of days_since_prior_order column
ords_prods_merge['median_freq'] = ords_prods_merge.groupby(['user_id'])['days_since_prior_order'].transform(np.median)
```

```python
#Order Frequency Flag for Non-Frequent customer
ords_prods_merge.loc[ords_prods_merge['median_freq'] > 20, 'order_freq_flag'] = 'Non-Frequent customer'
```

```python
#Order Frequency Flag for Regular customer
ords_prods_merge.loc[(ords_prods_merge['median_freq'] > 10) & (ords_prods_merge['median_freq'] <= 20), 'order_freq_flag'] = 'Regular customer'
```

```python
#Order Frequency Flag for Frequent customer
ords_prods_merge.loc[ords_prods_merge['median_freq'] <= 10, 'order_freq_flag'] = 'Frequent customer'
```

```python
ords_prods_merge.head(20)
```

```python
#Checking the count for order frequency
ords_prods_merge['order_freq_flag'].value_counts(dropna = False)
```

```python
#Checking the count to ensure no empty entries
ords_prods_merge['order_freq_flag'].value_counts(dropna = False).sum()
```
